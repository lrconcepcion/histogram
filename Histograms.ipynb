```python
#Import matplotlib, numpy and, if using Jupyter notebooks, make sure to use inline
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
#Define a mean and a standard deviation, then use a numpy function to create an array of random numbers 
#with a normal distribution (i.e. 200 numbers).
mu = 80
sigma = 7
x = np.random.normal(mu, sigma, size=200)
```

```python
#Use plt.hist() to plot the histogram along with some keyword arguments
plt.hist(x, 20,
         density=True,
         histtype='bar',
         facecolor='b',
         alpha=0.5)

plt.show()
```

```python
#Answering real life analytical problems with histograms - first let's import numpy and pandas
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
%matplotlib inline
```

```python
#Make sure you install xlrd module using conda install -c anaconda xlrd --yes
#Then download the dataset Canadian Immigration available at 
# https://ibm.box.com/shared/static/lw190pt9zpy5bd1ptyg2aw15awomz9pu.xlsx 
df_can = pd.read_excel('Canada.xlsx')
df_can.head() #Take a look at the first 5 items on the dataset
```

```python
#Set the country name(OdName) as index - useful for quickly looking up countries using .loc method.
df_can.set_index('OdName', inplace=True)

# let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python
#Add total column
df_can['Total'] = df_can.sum(axis=1)

# let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python
print ('data dimensions:', df_can.shape)
```

```python
# finally, let's create a list of years from 1980 - 2013
# this will come in handy when we start plotting the data
years = list(map(str, range(1980, 2014)))

years
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
# let's quickly view the 2013 data
df_can[2013].head()
```

```python
# np.histogram returns 2 values
count, bin_edges = np.histogram(df_can[2013])

print(count) # frequency count
print(bin_edges) # bin ranges, default = 10 bins
```

```python
df_can[2013].plot(kind='hist', figsize=(8, 5))

plt.title('Histogram of Immigration from 195 Countries in 2013') # add a title to the histogram
plt.ylabel('Number of Countries') # add y-label
plt.xlabel('Number of Immigrants') # add x-label

plt.show()
```

```python
# 'bin_edges' is a list of bin intervals
count, bin_edges = np.histogram(df_can[2013])

df_can[2013].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Immigration from 195 countries in 2013') # add a title to the histogram
plt.ylabel('Number of Countries') # add y-label
plt.xlabel('Number of Immigrants') # add x-label

plt.show()
```

### This histogram is skewed right. Most countries that have migrated to Canada are providing between 0 and 5 thousand immigrants each.  A few countries have provided between 10 and 15 thousand migrants, and a handful of countries have high migration rates of 30 to 35 thousand migrants.
